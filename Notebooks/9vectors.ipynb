```python
import math
import matplotlib.pyplot as plt

def calculate_barycenter(CH4_percent, C2H6_percent, H2_percent, C2H2_percent, C2H4_percent):
    # Define center coordinates of the pentagon
    center_x = 0
    center_y = 0

    # Define the radius of the pentagon
    radius = 50

    # Calculate coordinates for the vertices A, B, C, D, and E
    angle_deg = 90  # Angle for the first vertex A
    vertices = {}
    for vertex in ['A', 'B', 'C', 'D', 'E']:
        angle_rad = math.radians(angle_deg)
        x = center_x + radius * math.cos(angle_rad)
        y = center_y + radius * math.sin(angle_rad)
        vertices[vertex] = (x, y)
        angle_deg += 72  # Increase angle by 72 degrees for the next vertex

    # Calculate gas points based on percentages
    CH4_point = ((100 - CH4_percent) / 100 * vertices['C'][0] + CH4_percent / 100 * vertices['B'][0],
                 (100 - CH4_percent) / 100 * vertices['C'][1] + CH4_percent / 100 * vertices['B'][1])

    C2H6_point = ((100 - C2H6_percent) / 100 * vertices['B'][0] + C2H6_percent / 100 * vertices['A'][0],
                  (100 - C2H6_percent) / 100 * vertices['B'][1] + C2H6_percent / 100 * vertices['A'][1])

    H2_point = ((100 - H2_percent) / 100 * vertices['A'][0] + H2_percent / 100 * vertices['E'][0],
                (100 - H2_percent) / 100 * vertices['A'][1] + H2_percent / 100 * vertices['E'][1])

    C2H2_point = ((100 - C2H2_percent) / 100 * vertices['E'][0] + C2H2_percent / 100 * vertices['D'][0],
                  (100 - C2H2_percent) / 100 * vertices['E'][1] + C2H2_percent / 100 * vertices['D'][1])

    C2H4_point = ((100 - C2H4_percent) / 100 * vertices['D'][0] + C2H4_percent / 100 * vertices['C'][0],
                  (100 - C2H4_percent) / 100 * vertices['D'][1] + C2H4_percent / 100 * vertices['C'][1])

    # Calculate the barycenter using gas points
    barycenter_x = (CH4_point[0] + C2H6_point[0] + H2_point[0] + C2H2_point[0] + C2H4_point[0]) / 5
    barycenter_y = (CH4_point[1] + C2H6_point[1] + H2_point[1] + C2H2_point[1] + C2H4_point[1]) / 5

    return (barycenter_x, barycenter_y), vertices, CH4_point, C2H6_point, H2_point, C2H2_point, C2H4_point 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the calculate_barycenter function
def calculate_barycenter(CH4_percent, C2H6_percent, H2_percent, C2H2_percent, C2H4_percent):
    # Define vertices of the pentagon
    vertices = {
        'H2': np.array([1, 0]),
        'C2H6': np.array([np.cos(2 * np.pi / 5), np.sin(2 * np.pi / 5)]),
        'CH4': np.array([np.cos(4 * np.pi / 5), np.sin(4 * np.pi / 5)]),
        'C2H4': np.array([np.cos(6 * np.pi / 5), np.sin(6 * np.pi / 5)]),
        'C2H2': np.array([np.cos(8 * np.pi / 5), np.sin(8 * np.pi / 5)])
    }

    # Rotate the vertices so that H2 is at the top
    rotation_matrix = np.array([[np.cos(np.pi/2), -np.sin(np.pi/2)], [np.sin(np.pi/2), np.cos(np.pi/2)]])
    for key in vertices:
        vertices[key] = rotation_matrix @ vertices[key]

    # Center of the pentagon
    center = np.array([0, 0])

    # Calculate points along the lines from the center to each vertex
    CH4_point = center + CH4_percent / 100 * vertices['CH4']
    C2H6_point = center + C2H6_percent / 100 * vertices['C2H6']
    H2_point = center + H2_percent / 100 * vertices['H2']
    C2H2_point = center + C2H2_percent / 100 * vertices['C2H2']
    C2H4_point = center + C2H4_percent / 100 * vertices['C2H4']

    points = [C2H2_point, H2_point, C2H6_point, CH4_point, C2H4_point]

    # Calculate barycenter of the new pentagon
    barycenter = np.mean(points, axis=0)

    return barycenter, vertices, CH4_point, C2H6_point, H2_point, C2H2_point, C2H4_point

# Given percentages
C2H4_percent = 26
CH4_percent = 55
C2H6_percent = 52
H2_percent = 80
C2H2_percent = 15


# Calculate the barycenter and points
barycenter, vertices, CH4_point, C2H6_point, H2_point, C2H2_point, C2H4_point = calculate_barycenter(CH4_percent, C2H6_percent, H2_percent, C2H2_percent, C2H4_percent)

# Plot the pentagon, barycenter, and gas percentage points
fig, ax = plt.subplots()

# Scale up the size of the plot
scale = 10
for key in vertices:
    vertices[key] = vertices[key] * scale

# Plot original pentagon vertices and connect them
for i in range(5):
    start = list(vertices.values())[i]
    end = list(vertices.values())[(i + 1) % 5]
    ax.plot([start[0], end[0]], [start[1], end[1]], 'b-')  # Draw lines connecting vertices
    ax.plot(start[0], start[1], 'bo')  # Plot vertices
    ax.text(start[0], start[1], list(vertices.keys())[i], fontsize=8, ha='right')  # Label vertices

# Plot the new pentagon points
points = [C2H4_point * scale, CH4_point * scale, C2H6_point * scale, H2_point * scale, C2H2_point * scale]
for point in points: # Plot each point
    ax.plot(point[0], point[1], 'go')  # Plot each point 

# Connect the new pentagon points
for i in range(5):
    start = points[i]
    end = points[(i + 1) % 5]
    ax.plot([start[0], end[0]], [start[1], end[1]], 'r--')  # Draw lines connecting points

# Plot the barycenter of the new pentagon
barycenter = barycenter * scale
ax.plot(barycenter[0], barycenter[1], 'ro')  # Plot the barycenter

# Plot lines from center to vertices of the original pentagon
center = np.array([0, 0])
for vertex in vertices.values():
    ax.plot([center[0], vertex[0]], [center[1], vertex[1]], 'k--')  # Draw lines from center to vertices

ax.set_aspect('equal', adjustable='box')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Pentagon and Barycenter')
plt.grid(True)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

df_Pentagon_test = df_test.copy() # copy the data to another dataframe
df_Pentagon_train = df_train.copy() # copy the data to another dataframe

# Change the order of columns
df_Pentagon_test = df_Pentagon_test.reindex(columns=['CH4', 'C2H6', 'H2', 'C2H2', 'C2H4']) 
df_Pentagon_train = df_Pentagon_train.reindex(columns=['CH4', 'C2H6', 'H2', 'C2H2', 'C2H4']) 

def calculate_barycenter(CH4_percent, C2H6_percent, H2_percent, C2H2_percent, C2H4_percent):
    # Define center coordinates of the pentagon
    center_x = 0
    center_y = 0

    # Define the radius of the pentagon
    radius = 100

    # Calculate coordinates for the vertices A, B, C, D, and E
    angle_deg = 90  # Angle for the first vertex A
    vertices = {} # Create an empty dictionary to store the vertices 
    for vertex in ['A', 'B', 'C', 'D', 'E']: 
        angle_rad = math.radians(angle_deg) # Convert angle to radians 
        x = center_x + radius * math.cos(angle_rad)
        y = center_y + radius * math.sin(angle_rad)
        vertices[vertex] = (x, y)
        angle_deg += 72  # Increase angle by 72 degrees for the next vertex 
    # this code is very smart 

    # Calculate gas points based on percentages
    CH4_point = ((100 - CH4_percent) / 100 * vertices['C'][0] + CH4_percent / 100 * vertices['B'][0],
                 (100 - CH4_percent) / 100 * vertices['C'][1] + CH4_percent / 100 * vertices['B'][1])

    C2H6_point = ((100 - C2H6_percent) / 100 * vertices['B'][0] + C2H6_percent / 100 * vertices['A'][0],
                  (100 - C2H6_percent) / 100 * vertices['B'][1] + C2H6_percent / 100 * vertices['A'][1])

    H2_point = ((100 - H2_percent) / 100 * vertices['A'][0] + H2_percent / 100 * vertices['E'][0],
                (100 - H2_percent) / 100 * vertices['A'][1] + H2_percent / 100 * vertices['E'][1])

    C2H2_point = ((100 - C2H2_percent) / 100 * vertices['E'][0] + C2H2_percent / 100 * vertices['D'][0],
                  (100 - C2H2_percent) / 100 * vertices['E'][1] + C2H2_percent / 100 * vertices['D'][1])

    C2H4_point = ((100 - C2H4_percent) / 100 * vertices['D'][0] + C2H4_percent / 100 * vertices['C'][0],
                  (100 - C2H4_percent) / 100 * vertices['D'][1] + C2H4_percent / 100 * vertices['C'][1])

    # Calculate the barycenter using gas points
    barycenter_x = (CH4_point[0] + C2H6_point[0] + H2_point[0] + C2H2_point[0] + C2H4_point[0]) / 5
    barycenter_y = (CH4_point[1] + C2H6_point[1] + H2_point[1] + C2H2_point[1] + C2H4_point[1]) / 5

    return barycenter_x, barycenter_y  

P_row_sum_test = df_Pentagon_test.sum(axis=1) # sum the rows
df_Pen_test = df_Pentagon_test.div(P_row_sum_test, axis=0) * 100 # divide the data by the sum of the rows

P_row_sum_train = df_Pentagon_train.sum(axis=1) # sum the rows
df_Pen_train = df_Pentagon_train.div(P_row_sum_train, axis=0) * 100 # divide the data by the sum of the rows

df_Pen_test['Cpx'], df_Pen_test['Cpy'] = zip(*df_Pen_test.apply(lambda row: calculate_barycenter(row['CH4'], row['C2H6'], row['H2'], row['C2H2'], row['C2H4']), axis=1))
df_Pen_train['Cpx'], df_Pen_train['Cpy'] = zip(*df_Pen_train.apply(lambda row: calculate_barycenter(row['CH4'], row['C2H6'], row['H2'], row['C2H2'], row['C2H4']), axis=1))

df_Pen_test = df_Pen_test.drop(columns=['CH4','C2H6','H2','C2H2','C2H4'],axis=1) # drop the fault type column
df_Pen_train = df_Pen_train.drop(columns=['CH4','C2H6','H2','C2H2','C2H4'],axis=1) # drop the fault type column

df_Pen_test.head() # show the first 5 rows of the data
df_Pen_train.head() # show the first 5 rows of the data 

df_Pen_test.head() # show the first 5 rows of the data 
pd.DataFrame.to_csv(df_Pen_test, 'data_Pentagon_test.csv') # save the data to a csv file
pd.DataFrame.to_csv(df_Pen_train, 'data_Pentagon_train.csv') # save the data to a csv file
```
